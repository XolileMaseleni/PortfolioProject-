```python
#Library import

# Library imports
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

# Adjust the configuration of the plots we will create
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)

# Read in the data
df = pd.read_csv(r'C:\Users\User\OneDrive\Documents\Data analyst work\Portfolio\movies.csv')

# Looking at the data
df.head()

```

```python
#Looking for missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Looking at datd types 

df.dtypes
```

```python
#changing data type

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df
```

```python
#create correct year column

df['yearcorrect'] = df['released'].astype(str).str[:12]

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python


df_numerized = df.copy()

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()

```

```python
df
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add titles and labels
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
